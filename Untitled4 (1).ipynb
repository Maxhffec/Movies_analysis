```python
import pandas as pd
import numpy as np
import csv
import json
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
import requests
```

```python
df['IMDB_Rating']=pd.to_numeric(df['IMDB_Rating'])
```

```python
df['Meta_score']=pd.to_numeric(df['Meta_score'])
```

```python
df=df.fillna(0)
df
```

```python
df.iloc[20]
```

```python
df['Meta_score_10']=df['Meta_score']/10
```

```python
df['Total_score']=(df['IMDB_Rating']+df['Meta_score_10'])/2
```

```python
df.iloc[20]
```

```python
df.loc[df['Meta_score_10'] == 0, 'Total_score'] = df['IMDB_Rating']
```

```python
df.head(25)
```

```python
#Понять влияние количества голосов на рейтинг.
average_score_by_rating=df.groupby('No_of_Votes')['Total_score'].mean().reset_index()
```

```python
# Построение линейного графика среднего рейтинга фильмов по годам
plt.figure(figsize=(10, 6))
plt.plot(average_score_by_rating['No_of_Votes'], average_score_by_rating['Total_score'], marker='o', linestyle='-')
plt.title('Средний рейтинг по количеству голосов')
plt.xlabel('Количество голосов')
plt.ylabel('Средний рейтинг')
plt.xlim(0, 2343110)
plt.grid(True)
plt.show()
```
